```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
# names=True means expect column names as the first line
# dtype=None means let numpy figure out the dataype
# notice in the new file all the comments were taken out
# the comments were confusing numpy, they may also have been removed with a 
# skiplines keyword, but John was tired of playing with numpy
data = np.genfromtxt("hlist_1.00000.names", names=True, dtype=None)
# mvir = data[:,10]
# rvir = data[:,11]
# Nproj = data[:,4]
# vmax = data[:,16]
# vpeak = data[:,37]
# x = data[:,17]
# y = data[:,18]
# z = data[:,19]
data.dtype
```

```python
# Changing the size of the plot
fig = plt.figure(figsize=(20,10))

ax = fig.add_subplot(111)

ax.scatter(data['mvir'],data['vmax'],s=5,c='k',alpha=0.3)
ax.set_xscale('log')
ax.set_yscale('log')

ax.set_xlabel("$M_{vir}$ $(M_{\odot})$",size=20)
ax.set_ylabel("$v_{max}$ $(km/s)$",size=20)
plt.show()
```

```python
# Changing the size of the plot
fig = plt.figure(figsize=(20,10))

ax = fig.add_subplot(111)

image = ax.hexbin(data['mvir'], data['vmax'], cmap="jet", bins='log', gridsize=50, xscale='log', yscale='log')

bar = fig.colorbar(image,orientation="vertical",pad=0.01)
bar.set_label('$log10(N+1)$',size=15)

ax.set_xlabel("$M_{vir}$ $(M_{\odot})$",size=20)
ax.set_ylabel("$v_{max}$ $(km/s)$",size=20)

plt.show()
```

```python
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111, projection='3d')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
print("original: ", X.shape, Y.shape)
X, Y = np.meshgrid(X, Y)
print("after meshgrid: ", X.shape, Y.shape)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.scatter(X, Y, Z)  # , alpha=0.3)

plt.show()
```

```python
# Changing the size of the plot
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111, projection='3d')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

Z[25:30, 25:30] += .2

R = np.sqrt(X[1:20, 10:20]**2 + Y[1:20, 10:20]**2)
Z[1:20, 10:20] += np.cos(R)

ax.plot_surface(X, Y, Z)  # , alpha=0.3)

plt.show()
```

```python
# Changing the size of the plot
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111)

contour = ax.contour(X, Y, Z, cmap="Greys") # cool, plasma

plt.show()
```

```python
# Changing the size of the plot
fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111)

contour = ax.contourf(X, Y, Z,cmap="viridis")

plt.show()
```
